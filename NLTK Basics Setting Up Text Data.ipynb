```python
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk.probability import FreqDist
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
```

```python
nltk.download('inaugural')
speechObama = nltk.corpus.inaugural.words('2009-Obama.txt')
```

```python
stop_words=set(stopwords.words("english"))
punc = [' ', ',', '.', ';', "'", '--', '-', ':']

#I've created a list of my own stop words to add to our list
#If you're working on different data and find any inconsistencies
#such as stray letters or punctuation, you can handle them this way

for x in punc:
    stop_words.add(x)
stopless_speech = []
for word in speechObama:
    if word not in stop_words:
        stopless_speech.append(word)
print(stopless_speech[:10])
```

```python
fdist = FreqDist(stopless_speech)
print(fdist)
fdist.most_common(5)
fdist.plot(30,cumulative=False)
plt.show()
```

```python
nltk.download('wordnet')
fix = WordNetLemmatizer()
lemma_speech = []
for x in stopless_speech:
    lemma_speech.append(fix.lemmatize(x,'v'))
fdist = FreqDist(lemma_speech)
print(fdist)
fdist.most_common(5)
fdist.plot(30,cumulative=False)
plt.show()
```
